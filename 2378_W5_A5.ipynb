```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulating daily call distributions
N = 1000  # Number of days to simulate
calls_per_day = np.random.randint(0, 8, size=N)  # Random calls between 0 and 7

# Compute PMF (Probability Mass Function)
unique, counts = np.unique(calls_per_day, return_counts=True)
pmf = counts / N

# Compute CDF (Cumulative Distribution Function)
cdf = np.cumsum(pmf)

# Plot PMF
plt.figure(figsize=(10, 5))
sns.barplot(x=unique, y=pmf, color='blue', alpha=0.7)
plt.xlabel("Number of Calls")
plt.ylabel("PMF")
plt.title("Probability Mass Function (PMF) of Calls Per Day")
plt.show()

# Plot CDF
plt.figure(figsize=(10, 5))
sns.lineplot(x=unique, y=cdf, marker='o', color='red')
plt.xlabel("Number of Calls")
plt.ylabel("CDF")
plt.title("Cumulative Distribution Function (CDF) of Calls Per Day")
plt.show()

```

```python
from scipy.stats import binom

# Hospital Emergency Cases
N_patients = 100  # Number of patients per day
p_surgery = 0.25  # Probability of requiring surgery

# i) Probability that exactly 20 patients need surgery
prob_20 = binom.pmf(20, N_patients, p_surgery)
print(f"Probability that exactly 20 patients need surgery: {prob_20:.4f}")

# ii) Probability that at least 30 patients need surgery
prob_30_or_more = 1 - binom.cdf(29, N_patients, p_surgery)
print(f"Probability that at least 30 patients need surgery: {prob_30_or_more:.4f}")

# iii) Expected number of patients needing surgery
expected_surgeries = N_patients * p_surgery
print(f"Expected number of patients requiring surgery: {expected_surgeries}")
```
